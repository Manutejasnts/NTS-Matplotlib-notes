# 2) Matplotlib (Title, Label, grid and subplot)

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,222,240,230,260,280,320,370,320,300])
font1 = {'family':'serif','color':'blue','size':25}
font2 = {'family':'serif','color':'black','size':15}
plt.title('NTS Data analysis',fontdict = font1,loc = 'left')
plt.xlabel('Average Pulse',fontdict=font2)
plt.ylabel('Calories',fontdict=font2)
plt.plot(x,y)
plt.show()
```

# Styling the subplot

```python
import numpy as np
import matplotlib.pyplot as plt
#plot1
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,222,240,230,260,280,320,370,320,300])
font1 = {'family':'serif','color':'red','size':25}
font2 = {'family':'serif','color':'black','size':15}
plt.subplot(1,2,1)
plt.plot(x,y,marker='s',mfc='b',mec='r',ms=5,linestyle='dashed',linewidth=0.5,color='c')
plt.title('NTS 1',fontdict=font1)
plt.xlabel("Displacment",fontdict=font2)
plt.ylabel("Time",fontdict=font2)
plt.grid()

#plot2
x = np.array([80,85,90,100,105,120,125])
y = np.array([240,222,300,260,280,350,300])
font1 = {'family':'serif','color':'red','size':25}
font2 = {'family':'serif','color':'black','size':15}
plt.subplot(1,2,2)
plt.plot(x,y,marker='o',mfc='b',mec='r',ms=5,linestyle='dashed',linewidth=0.5,color='c')
plt.title('NTS 2',fontdict=font1)
plt.xlabel("Speed",fontdict=font2)
plt.grid()


plt.suptitle('NTS Reading',fontdict=font1)
plt.show()
```
