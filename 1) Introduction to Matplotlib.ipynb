# Introduction to matplotlib

A low level graph plotting libary in python that serves as a visualization utility

```python
import matplotlib
print(matplotlib.__version__)
```

Plotting a graph using array

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([1,3,7,9,15])                      #parameter1
y1 = np.array([10,30,60,90,100])                 #parameter2
plt.plot(x1,y1)
plt.show()
```

Here Parameter1 is an array containing the point on x-axis

Here Parameter2 is an array containing the point on y-axis 

# Plotting without line

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([1,3,7,9,15])                      #parameter1
y1 = np.array([10,30,60,90,100])                 #parameter2
plt.plot(x1,y1,'o')
plt.show()
```

Default x point 

if we don't specify the point in the x-axis, they will get the default values (0,1,2,3,4,....) depending on the length of y-axis

```python
import numpy as np
import matplotlib.pyplot as plt                      
y1 = np.array([10,30,60,90,100])                 
plt.plot(y1)
plt.show()
```

# Markers in matplotlib

Types of marker

You can choose any of these markers:

Marker Description

'o' --> 'Circle'

'*' --> 'Star'

'.' --> 'Point'

',' --> 'Pixel'

'x' --> 'X'

'X' --> 'X (filled)

'+' --> 'Plus'

'P' --> 'Plus (filled)'

's' --> 'Square'

'D' --> 'Diamond'

'd' --> 'Diamond (thin)'

'p' --> 'Pentagon'

'H' --> 'Hexagon'

'h' --> 'Hexagon'

'v' --> 'Triangle Down'

'^' --> 'Triangle Up'

'<' --> 'Triangle Left'

'>' --> 'Triangle Right'

'1' --> 'Tri Down'

'2' --> 'Tri Up'

'3' --> 'Tri Left'

'4' --> 'Tri Right'

'|' --> 'Vline'

'_' --> 'Hline'

# Format string (marker line color)

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([10,20,30,40,50])                      
y1 = np.array([20,60,30,100,40])                 
plt.plot(x1,y1,'o:r')
plt.show()
```

# Line Reference

Line Syntax Description

'-'  -->'Solid line'

':'  -->'Dotted line'

'--' -->'Dashed line '

'-.' -->'dashdot line'



# Color Reference

Color Syntax Description

'r' Red

'g' Green

'b' Blue

'c' Cyan

'm' Magenta

'y' Yellow

'k' Black

'w' White

# Marker size (ms)

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([10,20,30,40,50])                      
y1 = np.array([20,60,30,100,40])                 
plt.plot(x1,y1,marker='*',ms=15)
plt.show()
```
