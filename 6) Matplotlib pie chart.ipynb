# Matplotlib pie chart

### Creating a pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
plt.pie(a)
plt.show()
```

### Labeling the pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
plt.pie(a,labels=la)
plt.show()
```

### Changing the start angle

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
plt.pie(a,labels=la,startangle=60)
plt.show()
```

### Changing the color of the pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
co = np.array(["red","green","pink","brown"])
plt.pie(a,labels=la,startangle=60,colors=co)
plt.show()
```

### Highlighting or Exploding the cor form the pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
exp = np.array([0.2,0,0,0])
plt.pie(a,labels=la,explode=exp)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
exp = np.array([0.2,0.1,0.2,0.1])
plt.pie(a,labels=la,explode=exp)
plt.show()
```

### To add shadow to pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
exp = np.array([0.2,0,0,0])
plt.pie(a,labels=la,explode=exp,shadow=True)
plt.show()
```

### Using legend

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
plt.pie(a,labels=la)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([35,25,25,15])
la = np.array(["Audi","BMW","Honda","TATA"])
exp = np.array([0.2,0,0,0])
plt.pie(a,labels=la,explode=exp)
plt.legend(title="color of the company",loc="center")
plt.show()
```
