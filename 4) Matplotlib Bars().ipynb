# Matplotlib Bars()

### 1)Creating bars

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.bar(a,b)
plt.show()
```

### 2)Horizontal bar

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.barh(a,b)
plt.show()
```

### 3) Changing the color of the bars

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.bar(a,b,color="hotpink")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.bar(a,b,color="#4CAF50")
plt.show()
```

### 4) Changing the width of the bars

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.bar(a,b,width=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array(["A","B","C","D","E"])
b = np.array([20,40,30,60,40])
plt.barh(a,b,height=0.5)
plt.show()
```
