# 5) Matplotlib Histograms

### There are 250 students and after evaluting we got mean as 170 and standard deviation as 10. Let's plot a histogram for the above given information

```python
import numpy as np
x = np.random.normal(170,10,250)                   #just
print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title('Height of the students')
plt.hist(x)
plt.show()
```
